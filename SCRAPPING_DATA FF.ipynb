```python
from google_play_scraper import app

import pandas as pd

import numpy as np
```

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.dts.freefireth',
    lang='id', 
    country='id', 
    sort=Sort.MOST_RELEVANT, 
    count=5000,
    filter_score_with=None
)
```

```python
df_FF = pd.DataFrame(np.array(result),columns=['review'])

df_FF = df_FF.join(pd.DataFrame(df_FF.pop('review').tolist()))

df_FF.head()
```

```python
df_FF[['userName', 'score','at', 'content']].head()
```

```python
new_df = df_FF[['userName', 'score','at', 'content']]
sorted_df = new_df.sort_values(by='at', ascending=False)
sorted_df.head()
```

```python
my_df = sorted_df[['userName', 'score','at', 'content']] #get userName, rating, date-time, and reviews only
```
