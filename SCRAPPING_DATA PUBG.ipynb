```python
from google_play_scraper import app

import pandas as pd

import numpy as np
```

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.tencent.ig',
    lang='id', 
    country='id',
    sort=Sort.MOST_RELEVANT, 
    count=5000, 
    filter_score_with=None 
)
```

```python
df_PUBG = pd.DataFrame(np.array(result),columns=['review'])

df_PUBG = df_PUBG.join(pd.DataFrame(df_PUBG.pop('review').tolist()))

df_PUBG.head()
```

```python
df_PUBG[['userName', 'score','at', 'content']].head()  
```

```python
new_df = df_PUBG[['userName', 'score','at', 'content']]
sorted_df = new_df.sort_values(by='at', ascending=False)
sorted_df.head()
```

```python
my_df = sorted_df[['userName', 'score','at', 'content']] 
```

```python
my_df.head()
```
