## Melihat banyak User yang memberikan rating pada tahun 2023

## Dataset COD

```python
from datetime import datetime
import calendar
```

```python
# Filter dataset untuk tahun 2023 dan bulan 1-6
filtered_data = df_COD[(df_COD['tahun'] == 2023) & (df_COD['bulan'].between(1, 6))].copy()

# Mengubah nilai bulan menjadi nama bulan dalam urutan yang benar
filtered_data.loc[:, 'Bulan'] = filtered_data['bulan'].apply(lambda x: calendar.month_name[x])

# Membuat tabel frekuensi rating berdasarkan bulan
rating_counts_COD = filtered_data.groupby('bulan')['Rating'].value_counts().unstack().fillna(0)

# Mengubah label rating
rating_counts_COD.columns = ['Rating ' + str(rating) for rating in rating_counts_COD.columns]

# Mereset indeks
rating_counts_COD = rating_counts_COD.reset_index()

# Menampilkan hasil
print(rating_counts_COD)
```

## Dataset FF

```python
# Filter dataset untuk tahun 2023 dan bulan 1-6
filtered_data = df_FF[(df_FF['tahun'] == 2023) & (df_FF['bulan'].between(1, 6))].copy()

# Mengubah nilai bulan menjadi nama bulan dalam urutan yang benar
filtered_data.loc[:, 'Bulan'] = filtered_data['bulan'].apply(lambda x: calendar.month_name[x])

# Membuat tabel frekuensi rating berdasarkan bulan
rating_counts_FF = filtered_data.groupby('bulan')['Rating'].value_counts().unstack().fillna(0)

# Mengubah label rating
rating_counts_FF.columns = ['Rating ' + str(rating) for rating in rating_counts_FF.columns]

# Mereset indeks
rating_counts_FF = rating_counts_FF.reset_index()

# Menampilkan hasil
print(rating_counts_FF)
```

```python
# Filter dataset untuk tahun 2023 dan bulan 1-6
filtered_data = df_PUBG[(df_PUBG['tahun'] == 2023) & (df_PUBG['bulan'].between(1, 6))].copy()

# Mengubah nilai bulan menjadi nama bulan dalam urutan yang benar
filtered_data.loc[:, 'Bulan'] = filtered_data['bulan'].apply(lambda x: calendar.month_name[x])

# Membuat tabel frekuensi rating berdasarkan bulan
rating_counts_PUBG = filtered_data.groupby('bulan')['Rating'].value_counts().unstack().fillna(0)

# Mengubah label rating
rating_counts_PUBG.columns = ['Rating ' + str(rating) for rating in rating_counts_PUBG.columns]

# Mereset indeks
rating_counts_PUBG = rating_counts_PUBG.reset_index()

# Menampilkan hasil
print(rating_counts_PUBG)
```

## Menghitung korelasi dari pemberian Rating pada dataset yang satu dengan pemberian rating pada dataset yang lainnya

```python
datasets = [rating_counts_COD, rating_counts_FF, rating_counts_PUBG]
dataset_names = ['COD', 'FF', 'PUBG']

for i in range(len(datasets)):
    for j in range(i+1, len(datasets)):
        dataset1 = datasets[i]
        dataset2 = datasets[j]
        dataset_name1 = dataset_names[i]
        dataset_name2 = dataset_names[j]
        
        print(f"Korelasi antara Rating pada Dataset {dataset_name1} dengan Dataset {dataset_name2}:")
        
        for rating in range(1, 6):
            correlation = dataset1[f'Rating {rating}.0'].corr(dataset2[f'Rating {rating}.0'])
            print(f"Korelasi antara Rating {rating} pada Dataset {dataset_name1} dengan Rating {rating} pada Dataset {dataset_name2}: {correlation}")
        
        print()
```
